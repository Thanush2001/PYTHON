# PRAMAPADHAM GAME

```python
import time
import sys
import random

snakes={
    17:7,
    54:34,
    62:19,
    64:60,
    87:36,
    95:75,
    93:73,
    98:79}

ladders={
    4:14,
    9:31,
    1:38,
    21:42,
    28:84,
    51:67,
    72:91,
    80:99}


max=100
dice_faces=6


def diceval():
    return random.randint(1,dice_faces)

def snakebit(user,currentpos,oldpos):
    print("Yup!snake was bit "+user+" moved from "+str(currentpos)+" to "+str(oldpos))

def gotladder(user,currentpos,oldpos):
    print("ohoo! got ladder "+user+" moved from "+str(currentpos)+" to "+str(oldpos))
    
def moveuser(user,dicevalue,current_pos):
    finalval=current_pos+dicevalue
    print(user," move from "+str(current_pos)+" to "+str(finalval))

    
    if finalval>max:
        print("you need",finalval-current_pos,"to win this game! please try again.")
        return current_pos

    if finalval in snakes:
          oldposition=snakes.get(finalval)
          snakebit(user,finalval,oldposition)
    elif finalval in ladders:
        oldposition=ladders.get(finalval)
        gotladder(user,finalval,oldposition)
    else:
        oldposition=finalval
    return oldposition



def win(finalpos,user):
    if(finalpos==max):
        print("Congrats",user,"you won the match.")
        sys.exit(1)
   
def start():
    print("Welcome our snake and ladder game")
    user1=input("Enter the user1 name:")
    user2=input("Enter the user2 name:")
    position_user1=0
    position_user2=0
 
    while True:
        input(user1+" press enter the  roll dice")
        dice_value=diceval()
        print("dice value=",dice_value)
        finalpos=moveuser(user1,dice_value,position_user1)
        position_user1=finalpos
        win(position_user1,user1)

        input(user2+" press enter the  roll dice")
        dice_value=diceval()
        print("dice value=",dice_value)
        finalpos=moveuser(user2,dice_value,position_user2)
        position_user2=finalpos
        win(position_user2,user2)
    
if __name__=="__main__":
    start()
```
