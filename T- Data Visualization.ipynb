```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([20,40,60,80,100])
plt.plot(x,y,marker="*",)
plt.show()
```

```python
years=[2011,2012,2013,2014,2015]
yield_apples=[0.895,0.915,0.920,0.935,0.945]
plt.plot(years,yield_apples)
```

```python
plt.plot(years,yield_apples)
plt.xlabel("year")
plt.ylabel("yield(tons)")
```

```python
years=range(2000,2008)
yield_apples=[0.895,0.915,0,920,0.935,0.945,0.953,0.962,0.967]
yield_orange=[0.023,0.927,0.933,0.937,0.942,0.948,0.955,0.961]

plt.figure(figsize=(8,6))
plt.plot(years,yield_apples)
plt.plot(years,yield_orange)

plt.xlabel("year")
plt.ylabel("yield(tons)")

plt.title("Crop yield in state")
plt.legend(["apples","oranges"])

plt.scatter(yield_apples,years)
plt.scatter(yield_orange,years)

plt.xlabel("yield")
plt.ylabel("year")

plt.legend(["apples","orange"])
```

```python
names=['A','B','C']
values=[1,10,100]

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.scatter(names,values)
plt.suptitle('Categorical plotting')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("movies.csv")
data.head()
```

```python
# Scatter plot

plt.style.use("ggplot")
plt.figure(figsize=(8,6))
sns.regplot(x=data["Duration"],y=data["Rating"])
plt.show()
```

```python
data.shape
```

```python
sns.lmplot(x='Duration',y='Rating',data=data,
           fit_reg=False,
           hue='Vote',
           legend=True,
           palette="Set1",height=6)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.displot(data['Rating'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.displot(data['Rating'],kde=False,bins=5)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.displot(x="Duration",data=data)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.displot(x="Duration",data=data,
            hue="Gross")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(y=data["Gross"])
plt.show()
```
